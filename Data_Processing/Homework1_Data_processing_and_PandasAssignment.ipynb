```python
import pandas as pd
df = pd.read_csv('archive/water_potability.csv')
df
```

```python
# Kiểm tra giá trị bị thiếu
missing_values = df.isnull().sum()
missing_values
```

```python
# Điền giá trị bị thiếu bằng giá trị trung bình của cột
df['Sulfate'].fillna(df['Sulfate'].mean(), inplace=True)
df['Trihalomethanes'].fillna(df['Trihalomethanes'].mean(), inplace=True)
df['ph'].fillna(df['ph'].mean(), inplace=True)
missing_values = df.isnull().sum()
missing_values
```

```python
# Chuẩn hóa Min-Max để đưa dữ liệu về khoảng [0, 1]
data_normalized = (df - df.min()) / (df.max() - df.min())
data_normalized
```

```python
#VD về Data transformation
# Dữ liệu gốc
data = pd.DataFrame({"gender": ["nam", "nữ", "nam", "nữ", "nam", "nữ", "nam"]})

# Sử dụng hàm get_dummies của Pandas để mã hóa 'gender' thành các cột nhị phân 'gender_nữ' và 'gender_nam'
# Bên cạnh đó ta truyền vào tham số 'drop_first=True' để chỉ giữ lại một trong hai cột nhị phân để tránh hiện tượng đa cộng tuyến.
data_encoded = pd.get_dummies(data, columns=["gender"], drop_first=True)
data_encoded
```
